```python
a = np.linspace(0, 10, 11)
```

```python
b = a**4
```

```python
x = np.arange(0, 10)
```

```python
y = 2*x
```

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')


# SMALL AXES
axes2 = fig.add_axes([0.2, 0.7, 0.25, 0.25])
axes2.plot(a,b)
axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')


plt.show()
```

```python
type(fig)
```

```python
fig = plt.figure(figsize=(4,2), dpi=200)

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

fig.savefig('new_figure.png', bbox_inches='tight')
```

## Matplotlib subplot

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y LABEL 1,0')

axes[1][1].plot(a,b)
axes[1][1].set_title('TITLE 1,1')

fig.suptitle("Figure level", fontsize=16)

#plt.tight_layout()
fig.subplots_adjust(wspace=0.3, hspace=0.3)


fig.savefig('new_subplot.png', bbox_inches='tight')
```

### Legends

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label='X vs X')

ax.plot(x,x**2, label='X vs X**2')

ax.legend(loc=(0.5, 0.15)) #   (  )  
```

### Styling

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='red', marker='o', markerfacecolor='red', markeredgewidth=2, markeredgecolor='orange') # RGB HEX code

lines = ax.plot(x, x+1, color='#097a27', lw=3, ls='--', marker='s', ms=10) # you can code 'linewidth=3', linestyle='-.'
#ms == markersize

lines[0].set_dashes([3,2,3,2]) # 3 point line - 2 point white space + 3 point line + 2 point white space and so on.
```
