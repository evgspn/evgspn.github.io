```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/cancer_tumor_data_features.csv')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X.mean(axis=0)
```

```python
scaled_X - scaled_X.mean(axis=0)
```

```python
# scaled_X
```

```python
coveriance_matrix = np.cov(scaled_X, rowvar=False)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(coveriance_matrix)
```

```python
# eigen_vectors.shape
```

```python
# PCA 30 feature space --> 30 PC space --> 2
# How many PC to choose? N = 2 (or 3, but 2 easy to visualize)
```

```python
num_components = 2
```

```python
# sorted_keys = np.argsort()
```

```python
# [0,1,2]
np.argsort([2,1,3])
```

```python
# result upper:
# [1,2,3]
```

```python
sorted_keys = np.argsort(eigen_values)[::-1][:num_components]
```

```python
eigen_values, eigen_vectors = eigen_values[sorted_keys], eigen_vectors[:, sorted_keys]
```

```python
# Original data --> project --> eigen_vectors
principal_components = np.dot(scaled_X, eigen_vectors)
```

```python
from sklearn.decomposition import PCA
```

```python
pc_result = pca_model.fit_transform(scaled_df)
```

```python
pca_model.components_
```

```python
df_comp = pd.DataFrame(pca_model.components_, index=['PC1', 'PC2'], columns=df.columns)
```
