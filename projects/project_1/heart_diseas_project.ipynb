## Датасет, краткие сведения.

Датасет содержит 14 показателей, основанных на физическом тестировании пациентов. Берутся образцы крови, плюс пациент также проводит небольшой тест с физической нагрузкой. Поле «target» сообщает о наличии заболевания сердца (0 — отсутствие заболевания, 1 — наличие заболевания; другими словами: 0 - меньше шансов на сердечный приступ, 1 - больше шансов на сердечный приступ).

Информация о показателях (исходный источник - https://archive.ics.uci.edu/ml/datasets/Heart+Disease)

* age (возраст пациента)
* sex (пол пациента)
* cp (тип боли в груди, 4 значения) 
* trestbps (артериальное давление в состоянии покоя).
* chol (serum cholestoral in mg/dl (холестерин)
* fbs > 120 mg/dl (Уровень сахара в крови натощак)
* restecg (результаты электрокардиографии в покое (значения: 0,1,2))
* thalach (достигнутая максимальная частота сердечных сокращений)
* exang (стенокардия, вызванная физической нагрузкой)
* oldpeak = ST depression induced by exercise relative to rest (ST-депрессия, вызванная физическими упражнениями, по сравнению с отдыхом /знать бы смысл этой штуки! - тем интереснее выяснить значимость этого входного фактора./)
* slope (наклон пикового ST-сегмента при нагрузке)
* ca (количество магистральных сосудов (0-3), окрашенных флюороскопией)
* thal (результат стресс-теста с талием ~ (0,3))
* target (целевая переменная: 0 – отсутствие заболеваний сердца, 1 – наличие заболеваний сердца).

Чтобы эти единицы измерения перевести в ммоль/л надо _показатель в мг/дл разделить на 18_. То есть "меньше 120 мг/дл" равнозначен "меньше ~6,6 ммоль/л". Как видим, поедание любимых углеводов не говорит о том, что у вас будут проблемы с сердцем. Ровно как и наоборот.

# Машинное Обучение

## Train | Test Split and Scaling

**Разделим наш датасет на два объекта: X и y.

```python
X = df.drop('target',axis=1)
y = df['target']
```

**Применим из библиотеки любимого Scikit-learn инструмент "train test split" на датасет, тестовый размер из выборки составит 10%.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
# Проверим формы массивов разделенных данных (y = 10%): 
print("Форма X_train:     ", X_train.shape)
print("Форма X_test:      ", X_test.shape)
print("Форма y_train:     ", y_train.shape)
print("Форма y_test:      ", y_test.shape)
```

**Также используем StandardScaler для нормализации данных. Для того, чтобы избежать утечки данных из test-сета, применим метод fit только к training-данным:**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

## Логистическая Регрессия. Модель.

**Создадим модель логистической регрессии с Cross-Validation и найдем эффективные значения гиперпараметров (в частности, параметр _С_)**

```python
# Импортируем класс модели из семейства моделей:

from sklearn.linear_model import LogisticRegressionCV 
```

```python
# Создадим экземпляр модели:

log_model = LogisticRegressionCV()
```

```python
# Применим метод fit на тренировочные данные:

log_model.fit(scaled_X_train,y_train)
```

**Посмотрим, используемые параметры:** 

```python
log_model.C_
```

```python
log_model.get_params()
```

### Коэффициенты

**Посмотрим на коэффициенты модели:**

```python
log_model.coef_
```

**Предтавим полученные коэффициенты в графическом виде:**

```python
coefs = pd.Series(index=X.columns,data=log_model.coef_[0])
```

```python
# Отсортируем полученные коэффициенты от меньшего к большему:

coefs = coefs.sort_values()
```

```python
# Построим график:

plt.figure(figsize=(10,6))
sns.barplot(x=coefs.index,y=coefs.values);
```

---------

## Оценка Показателей Модели

**Оценим модель на 10% данных из test-сета.**

**Оценку будем производить на основе следующих метрик:**
* Confusion Matrix Array
* Confusion Matrix Plot
* Classification Report

```python
# Импортируем необходимые метрики:

from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix
```

```python
# прогнозирование значений:

y_pred = log_model.predict(scaled_X_test)
```

```python
# Отобразим матрицу ошибок:

confusion_matrix(y_test,y_pred)
```

```python
# Матрица ошибок в графическом виде:

plot_confusion_matrix(log_model,scaled_X_test,y_test);
```

Как видим, у нас только 5 ошибок. Для такого небольшого тренировочного сетка это можно считать вполне неплохим результатом. 

```python
# "Распечатаем" показатели измерения точности оценки произоведенных вычислений, 
# то есть качества прогнозов алгоритма классификации, используя classification_report:

print(classification_report(y_test,y_pred))
```

В целом, показатели вполне достаточные, чтобы судить о полученных результатах. То есть в ~80 процентов случаев полученная модель будет приносить кому-то пользу, поскольку поможет обнаружить болезни сердца по определенным физическим показателям, которые есть у пациента.
Разумеется, в зависимости от контекста ситуации - даже одна ошибка может быть "непозволительно большой".
